```python
from intensities import process_df
from meteo import daily_resume
import pandas as pd
```

```python
data = pd.read_csv('data/movilidad_barrios.csv')
intensities = process_df(data)
meteo = daily_resume()

df_combined = pd.merge(
    left=intensities, 
    right=meteo, 
    on='day', 
    how='left'
)

df_combined.to_csv('data/final_data.csv', index = False)
```
